```python
sns.boxplot(x='class', y= 'age', hue= 'sex', data = Titanic)
plt.ylim(0,100)
```

```python
sns.stripplot(x='class', y= 'age', hue= 'sex', data = Titanic, jitter = True, dodge=True)
```

```python
plt.figure(figsize = (15,8))
sns.swarmplot(x='class', y= 'age', hue= 'sex', data = Titanic, dodge=True, palette = 'Reds')
```

Pivot Table

```python
Titanic.pivot_table(index= 'who', columns = 'class' , values = 'fare')
```

# Cluster Map

PairPLot

```python
sns.pairplot(Titanic, kind = 'scatter',  palette = 'Blues')
```

PairGrid

```python
df = Titanic.drop(['adult_male', 'deck',  'embark_town', 'alive' ,  'alone','embarked', 'class'], axis =1)
print(df.head())
```

```python
pair = sns.PairGrid(df)
pair.map_diag(plt.hist)
pair.map_upper(sns.kdeplot)
pair.map_lower(plt.scatter)
```
